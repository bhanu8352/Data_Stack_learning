# *args, **kwargs 

are used to allow functions to accept a arbitrary number of arguments. These fetaures  provides  great flexibility 

```python
def fun(*args):
    return max(args)
print(fun(20,5))
```

**kwargs example - it collects keyword arguments into a dictionary


```python
def fun(**kwargs):
    for k, val in kwargs.items():
        print(k, val)

fun(a=1, b=4, c=6)
```

#reduce()

reduce function from functools applies a  function cumulatively to an iterable, reducing it into a single value. 

-functools - offersa colletcion of tools that simplify working with functions and callable objects. it includes uilities to modify, extend, or optimize fnctions without re writing the core logic. 

```python
from functools import reduce
def add(x, y):
    return x+y

a = [1,2,3,4,5]
res = reduce(add, a)

print(res)
```

```python
a = [1, 2,3,4,5]
res = reduce(lambda x,y : x*y, a, 10)
print(res)
```

```python
import functools
import operator

a = [1,2,3,4,5]
print(functools.reduce(operator.add, a))
```

#filter function 

-used to extract elements from an iterable (like a list or tuple or set) that satisfies a condition

syntax: filter(function, iterable)

```python
a = [1,2,3,4,5,6,7,8,9]
b = filter(lambda x: x%2 ==0, a)

c = map(lambda x: x*2, b)

print(list(c))
```
